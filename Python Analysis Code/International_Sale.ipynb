International Sales Report Analysis

```python
import pandas as pd

# Load dataset
df = pd.read_csv("C:/Users/Jitesh Jangam/Desktop/Future Int/Ecommerce/Dataset/International sale Report.csv")

# 1. Remove unwanted columns (if any)
df = df.loc[:, ~df.columns.str.contains("Unnamed|index", case=False)]

# 2. Convert DATE column
df["DATE"] = pd.to_datetime(df["DATE"], errors="coerce")

# 3. Convert numeric columns
df["PCS"] = pd.to_numeric(df["PCS"], errors="coerce")
df["RATE"] = pd.to_numeric(df["RATE"], errors="coerce")
df["GROSS AMT"] = pd.to_numeric(df["GROSS AMT"], errors="coerce")

# 4. Remove rows with missing critical values
df = df.dropna(subset=["DATE", "GROSS AMT"])

# Clean SKU column
df["SKU"] = df["SKU"].astype(str).str.strip().str.upper()

# Remove rows with missing SKU
df = df[df["SKU"] != ""]

# 5. Save cleaned file
df.to_csv("international_sales_cleaned.csv", index=False)

print("International sales data cleaned")
```
