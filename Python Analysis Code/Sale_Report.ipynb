Sale_Report Analysis and Cleaning

```python
import pandas as pd

# Load dataset
df = pd.read_csv("C:/Users/Jitesh Jangam/Desktop/Future Int/Ecommerce/Dataset/Sale Report.csv")

# 1. Drop useless index column
df = df.drop(columns=["index"])

# 2. Standardize column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace(".", "", regex=False)
)

# 3. Strip whitespace from all string columns
string_cols = df.select_dtypes(include="object").columns
df[string_cols] = df[string_cols].apply(lambda x: x.str.strip())

# 4. Convert stock to integer (it is quantity, not float)
df["stock"] = pd.to_numeric(df["stock"], errors="coerce").fillna(0).astype(int)

# 5. Remove duplicate SKUs (keep latest)
df = df.drop_duplicates(subset="sku_code", keep="last")

# 6. Validate data
assert df["stock"].ge(0).all(), "Negative stock found"
# assert df.isnull().sum().sum() == 0, "Null values still exist"

# 7. Save cleaned file
df.to_csv("Sale_Report_CLEAN.csv", index=False)

print("✅ Sale Report cleaned and saved successfully")
```
